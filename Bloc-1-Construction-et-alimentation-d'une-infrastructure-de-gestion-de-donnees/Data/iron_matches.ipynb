```python
from riotwatcher import LolWatcher, ApiError
import pandas as pd
import time

# golbal variables
api_key = 'RGAPI-99be67a3-9fd0-4d2f-bd38-3fe5f3c68d05'
watcher = LolWatcher(api_key)
region = 'euw1'
game_mode = "RANKED_SOLO_5x5"
iron = "IRON"
iron_division = "IV"
```

```python
iron_players = watcher.league.entries(region, game_mode, iron, iron_division)
```

```python
iron_players_lst = []

for i in iron_players:
    iron_players_lst.append(i["summonerId"])

iron_players_lst
```

```python
player_data = watcher.summoner.by_id(region, iron_players_lst[0])
player_data
```

```python
puuid_lst = []

for i in range(len(iron_players_lst)):
    time.sleep(1.3)
    puuid_lst.append(watcher.summoner.by_id(region, iron_players_lst[i])['puuid'])
    print (f"Players done so far: {i+1}/{len(iron_players_lst)}")

puuid_lst
```

```python
matches_lst = []
counter = 0

for i in puuid_lst:
    time.sleep(1.3)
    matches_lst.append(watcher.match.matchlist_by_puuid(region, i, 0, 100))
    counter+=1
    print (f"Players done so far: {counter}/{len(puuid_lst)}")


matches_lst = [x for xs in matches_lst for x in xs]
len(matches_lst)
matches_lst = list(set(matches_lst))
len(matches_lst)
```

```python
df2 = pd.DataFrame()
soloqueue_counter = 0
other_counter = 0
total_counter = 0

for i in matches_lst:
    time.sleep(1.3)
    total_counter+=1
    try:
        match_detail = watcher.match.by_id(region, i)
        # print (match_detail['info']["queueId"])
        if match_detail['info']["queueId"] == 420:
            participants = []
            soloqueue_counter +=1
            for row in match_detail['info']['participants']:
                participants_row = {}
                participants_row['champion'] = row['championName']
                participants_row['position'] = row['individualPosition']
                participants_row['win'] = row['win']
                participants_row['kills'] = row['kills']
                participants_row['deaths'] = row['deaths']
                participants_row['assists'] = row['assists']
                participants_row['totalDamageDealt'] = row['totalDamageDealt']
                participants_row['goldPerMinute'] = row["challenges"]['goldPerMinute']
                participants_row['champLevel'] = row['champLevel']
                participants_row['laneMinionsFirst10Minutes'] = row["challenges"]['laneMinionsFirst10Minutes']
                participants_row['jungleCsBefore10Minutes'] = row["challenges"]['jungleCsBefore10Minutes']
                participants_row['outerTurretExecutesBefore10Minutes'] = row["challenges"]['outerTurretExecutesBefore10Minutes']                
                participants_row['visionScore'] = row['visionScore']
                participants_row['visionScoreAdvantageLaneOpponent'] = row["challenges"]['visionScoreAdvantageLaneOpponent']
                participants_row['pentaKills'] = row['pentaKills']
                participants_row['perfectDragonSoulsTaken'] = row["challenges"]['perfectDragonSoulsTaken']
                participants_row['gamesNumber'] = soloqueue_counter
                participants_row['gamesId'] = i
                participants.append(participants_row)
            df = pd.DataFrame(participants)
            df2 = pd.concat([df2, df], axis=0)
        else:
            other_counter+=1
            pass    
    except:
        pass
    print (f"Check counter: {total_counter}/{len(matches_lst)} - {soloqueue_counter} ranked games saved - {other_counter} non-ranked games not saved.")

df2.to_csv('iron_ranked.csv')
```
